```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv("Dataset\\cleaned_title_netflix_dataset.csv")

df.head()
```

```python
df.info()
```

```python
for column in df.columns:
    df[column] = df[column].fillna("missing")
    
df.head()
```

```python
year = df["date_added"].str.split(pat=",", expand=True)[1]
month = df["date_added"].str.split(expand=True)[0]

year = year.apply(lambda x: x.strip())
month = month.apply(lambda x: x.strip())

df["month_added"] = month
df["year_added"] = year
```

```python
df.head()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])

df.head()
```

```python
df["seasonly"] = df["duration"].apply(lambda x: 1 if "Season" in x else 0)
df["duration_season"] = df["duration"].apply(lambda x: x.split()[0] if "Season" in x else 0)
df["minutely"] = df["duration"].apply(lambda x: 1 if "min" in x else 0)
df["duration_minute"] = df["duration"].apply(lambda x: x.split()[0] if "min" in x else 0)

df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 9))

descs = ""
for desc in df["description"]:
    descs += desc

words = WordCloud().generate(descs)

plt.imshow(words)
plt.axis("off")
plt.show()
```

```python
plots = pd.DataFrame(data={"family":np.zeros(df.shape[0]),
                           "life":np.zeros(df.shape[0]),
                           "friend":np.zeros(df.shape[0])})

for i in range(df.shape[0]):
    for feature in plots.columns:
        if feature in df["description"][i]:
            plots.loc[i, feature] = 1
        else:
            plots.loc[i, feature] = 0
            
df = pd.concat([df, plots], axis=1)
```
