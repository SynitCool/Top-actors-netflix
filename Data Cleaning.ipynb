# Data Cleaning

# Feature Information

### - show_id = Unique ID for every Movie / Tv Show
### - type = Identifier - A Movie or TV Show
### - title = Title of the Movie / Tv Show
### - director = Director of the Movie
### - cast = Actors involved in the movie / show
### - country = Country where the movie / show was produced
### - date_added = Date it was added on Netflix
### - release_year = Actual Release year of the move / show
### - rating = TV Rating of the movie / show
### - duration = Total Duration - in minutes or number of seasons
### - listed_in = Genre
### - description = The summary description

```python
import pandas as pd
import numpy as np
import datetime as dt

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset\\netflix_titles.csv")

df.head()
```

```python
df.info()
```

```python
missing_value = pd.DataFrame({"rate":np.zeros(df.shape[1])}, index=df.columns)

for column in df.columns:
    missing_value.loc[column] = round(df[column].isnull().sum() / df.shape[0] * 100, 1)
    
fig, ax = plt.subplots(figsize=(14, 5))

sns.barplot(data=missing_value, x=missing_value.index, y="rate")

missing_value[missing_value["rate"] != 0]
```

## Rating

```python
df[df["rating"].isnull()]
```

```python
df_tv = df[df["type"] == "TV Show"]
df_movie = df[df["type"] == "Movie"]
```

```python
df_listed_in_tv = pd.DataFrame({"genre":[],"count":[]})
df_listed_in_movie = pd.DataFrame({"genre":[],"count":[]})

listed_in_tv = []
listed_in_movie = []

for i in range(df_tv.shape[0]):
    listed_in_tv.extend(df_tv["listed_in"].iloc[i].split(","))
    
for i in range(len(listed_in_tv)):
    listed_in_tv[i] = listed_in_tv[i].strip()
    
for i in range(df_movie.shape[0]):
    listed_in_movie.extend(df_movie["listed_in"].iloc[i].split(","))
    
for i in range(len(listed_in_movie)):
    listed_in_movie[i] = listed_in_movie[i].strip()
```

```python
dict_tv = {}
dict_movie = {}

for i in listed_in_tv:
    dict_tv[i] = listed_in_tv.count(i)
    
for i in listed_in_movie:
    dict_movie[i] = listed_in_movie.count(i)
    
df_listed_in_tv["genre"] = list(dict_tv.keys())
df_listed_in_movie["genre"] = list(dict_movie.keys())

df_listed_in_tv["count"] = list(dict_tv.values())
df_listed_in_movie["count"] = list(dict_movie.values())
```

# Genre Conclusion

## TV Show :
  - The most common is International TV Shows
  
## Movies :
 - The most common is International Movies 

```python
listed_in_expand_tv = df_tv["listed_in"].str.split(pat=",", expand=True)
listed_in_expand_movie = df_movie["listed_in"].str.split(pat=",", expand=True)

listed_in_expand_tv.rename(columns={0:"genre_0", 1:"genre_1", 2:"genre_2"}, inplace=True)
listed_in_expand_movie.rename(columns={0:"genre_0", 1:"genre_1", 2:"genre_2"}, inplace=True)

for column in listed_in_expand_tv.columns:
    listed_in_expand_tv[column] = listed_in_expand_tv[column].apply(lambda x: x.strip() if pd.notnull(x) else x)
    listed_in_expand_movie[column] = listed_in_expand_movie[column].apply(lambda x: x.strip() if pd.notnull(x) else x)
```

```python
listed_in_expand_tv["International TV Shows"] = np.zeros(listed_in_expand_tv.shape[0])
listed_in_expand_movie["International Movies"] = np.zeros(listed_in_expand_movie.shape[0])

for column in listed_in_expand_tv.columns[:3]:
    listed_in_expand_tv.loc[listed_in_expand_tv[column] == "International TV Shows", "International TV Shows"] = 1
    listed_in_expand_movie.loc[listed_in_expand_movie[column] == "International Movies", "International Movies"] = 1
```

```python
df_tv["International TV Shows"] = listed_in_expand_tv["International TV Shows"]
df_movie["International Movies"] = listed_in_expand_movie["International Movies"]

df_tv.head()
```
